<b>üìò Pembahasan ‚Äì Operator dalam Python</b>

üîπ <b>Perbandingan</b> (hasilnya <code>True</code> / <code>False</code>)

- <code>>=</code> ‚Üí lebih besar atau sama dengan  
- <code><=</code> ‚Üí lebih kecil atau sama dengan

üîπ <b>Logika</b>

## ‚öúÔ∏èIf/Else

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
Pada dasarnya, pernyataan if-else üêç adalah salah satu konsep fundamental dalam pemrograman yang berfungsi untuk mengendalikan alur program. Bayangkan seperti sebuah persimpangan di jalan. Anda bisa memilih jalan mana yang akan diambil berdasarkan kondisi tertentu. if-else memungkinkan program Anda "memilih" jalan yang berbeda berdasarkan hasil dari suatu kondisi logis.

Bagaimana if-else Bekerja?
Setiap kali Anda menggunakan if, Python akan mengevaluasi ekspresi kondisional (seperti umur >= 18). Ekspresi ini akan selalu menghasilkan salah satu dari dua nilai: True (Benar) atau False (Salah).

Jika ekspresi menghasilkan True, blok kode di bawah if akan dieksekusi.

Jika ekspresi menghasilkan False, Python akan melompati blok kode if dan beralih ke blok else (jika ada).

Penting untuk diingat bahwa indentasi (spasi di awal baris) sangat krusial dalam Python. Blok kode yang termasuk dalam if atau else harus memiliki level indentasi yang sama, biasanya 4 spasi.
</div>

```python
suhu_air = 95

if suhu_air > 100:
    print("Airnya sudah mendidih!")
    print("Hati-hati, uapnya panas!")
else:
    print("Air belum mendidih.")
```

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
Pada contoh di atas, karena suhu_air adalah 95, yang mana tidak lebih besar dari 100, kondisi if menjadi False. Oleh karena itu, Python akan mengeksekusi blok kode di dalam else
</div>

```python
total = 18000

if total > 20000:
    print("Kamu mendapatkan diskon 20%")
elif total >= 13000:
    print("Kamu mendapatkan diskon 10%")
else:
    print("Kamu tidak mendapatkan diskon")
```

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
Seringkali, satu kondisi if-else tidak cukup. Di sinilah elif (singkatan dari "else if") berperan. elif memungkinkan Anda untuk memeriksa kondisi tambahan secara berurutan. Python akan mengecek setiap elif satu per satu jika kondisi sebelumnya salah.
</div>

```python
nilai = 85

if nilai >= 90:
    print("Nilai Anda A, luar biasa!")
elif nilai >= 80:  
    print("Nilai Anda B, bagus!")
elif nilai >= 70:
    print("Nilai Anda C, lumayan.")
else:
    print("Nilai Anda D, perlu belajar lebih giat.")
```

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
Dalam skenario ini, Python akan mengecek:

if nilai >= 90? False (karena 85 tidak lebih dari 90).

elif nilai >= 80? True (karena 85 lebih dari 80).

Karena kondisi elif ini True, Python akan menjalankan print("Nilai Anda B, bagus!") dan langsung keluar dari seluruh struktur if-elif-else ini. Kode di bawahnya (elif dan else yang lain) tidak akan pernah dieksekusi. Ini membuat alur program menjadi efisien.
</div>

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
Anda juga bisa membuat kondisi yang lebih kompleks menggunakan operator logis seperti and, or, dan not.

and: Kondisi menjadi True jika kedua sisi and adalah True.

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">

## ‚öúÔ∏èPengulangan

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
    <b>Apa Itu Perulangan (Looping)?</b> Secara sederhana, perulangan adalah instruksi yang meminta komputer untuk mengulang satu blok kode berulang kali. Bayangkan Anda harus menyalin 100 baris data dari satu file ke file lain. Menuliskan perintah salin sebanyak 100 kali adalah hal yang tidak praktis. Perulangan memungkinkan Anda untuk menuliskan perintah salin satu kali dan memberitahu program untuk mengulanginya 100 kali.

Di Python, ada dua jenis perulangan utama yang menjadi fondasi dari semua perulangan lainnya.
</div>

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
<b>1. Perulangan for</b> <br>
Perulangan for digunakan untuk iterasi yang pasti (definite iteration). Ini artinya, Anda tahu persis berapa kali perulangan akan berjalan, biasanya dengan mengulang setiap elemen dalam sebuah urutan (sequence). Urutan ini bisa berupa daftar, tuple, string, atau objek lain yang dapat diiterasi.
<br>
Iterasi di dalam Urutan
for sangat kuat karena ia akan secara otomatis mengambil setiap elemen dari urutan dan memberikannya ke variabel yang Anda tentukan.
</div>

<b>2. While</b>, jalan terus selama kondisi masih True. Perulangan while digunakan untuk iterasi yang tidak pasti (indefinite iteration). Perulangan ini akan terus berjalan selama sebuah kondisi bernilai True. Perulangan ini sangat berguna ketika Anda tidak tahu kapan perulangan akan berakhir.

Mengandalkan Kondisi Logis
Hal yang paling penting dari perulangan while adalah memastikan ada sesuatu di dalam blok kode yang akan membuat kondisi menjadi False di kemudian hari.

<br>

<b>Mengandalkan Kondisi Logis</b><br>
Hal yang paling penting dari perulangan while adalah memastikan ada sesuatu di dalam blok kode yang akan membuat kondisi menjadi False di kemudian hari.

```python
jumlah_tiket = 5
while jumlah_tiket > 0:
    print(f"Sisa tiket: {jumlah_tiket}. Penjualan sedang berlangsung.")
    jumlah_tiket -= 1

print("Tiket habis!")
```

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
Mengendalikan Alur Perulangan
Terkadang, Anda perlu mengubah alur perulangan dari dalam. Di sinilah break dan continue berperan.
<br>
<b>break</b>: Berfungsi sebagai "tombol keluar". Ini akan menghentikan seluruh perulangan secara total, bahkan jika kondisi while masih True atau masih ada elemen di for loop.
<br>
<b>continue</b>: Berfungsi sebagai "tombol lewati". Ini akan menghentikan iterasi saat ini dan langsung melompat ke iterasi berikutnya.

```python
for angka in range(1, 10):
    if angka == 3:
        continue
    if angka == 7:
        break
    print(f"Angka: {angka}")
```

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">

    String dianggap sebagai sequence atau rangkaian karakter.
    Setiap karakter memiliki posisi (index) dimulai dari 0 di sebelah kiri, dan bisa juga menggunakan index negatif dimulai dari -1 dari kanan.
    <br>
    Kapan String digunakan? String cocok digunakan ketika:
    <br>

    <br>
Menampilkan informasi ke layar.
    <br>

```python
## membalik string
s1 = "abcdefgh"
print(s1[::-1])

## menampilkan tiap 2 huruf
s2 = "1234567890"
for i in range(0,len(s2),2):
    print(s2[i:i+2])

## membalik tiap kata
s3 = "hello world python"
kata_list = s3.split()
kata_terbalik = [k[::-1] for k in kata_list]
print(" ".join(kata_terbalik))

## membuat vokal menjadi huruf besar
s4 = "openai is cool"
vokal = "aiueo"
s4_new = "".join([c.upper() if c in vokal else c for c in s4])
print(s4_new)

## menampilkan kata index ganjil
s5 = "one two three four five six"
kata_list5 = s5.split()
for i in range(len(kata_list5)):
    if i % 2 == 1:
        print(kata_list5[i])

## menampilkan tiap huruf
s6 = "python"
for huruf in s6:
    print(huruf)

## ganti spasi jadi underscore
s7 = "machine learning"
print(s7.replace(" ","_"))

## slicing ambil 3 huruf pertama
s8 = "abcde"
print(s8[:3])

## slicing ambil 2 huruf terakhir
s9 = "12345"
print(s9[-2:])

## print tiap huruf dipisah spasi
s10 = "data"
print(" ".join(s10))

## reverse tiap kata
s11 = "good morning everyone"
kata_list11 = s11.split()
kata_rev11 = [k[::-1] for k in kata_list11]
print(" ".join(kata_rev11))

## bikin huruf pertama tiap kata jadi besar
s12 = "hello world python"
print(" ".join([k.capitalize() for k in s12.split()]))

## ulang string 3 kali
s13 = "abc"
print(s13*3)

## ambil index genap
s14 = "abcdefgh"
print(s14[::2])

## ambil index ganjil
s15 = "abcdefgh"
print(s15[1::2])

## upper semua huruf
s16 = "python is fun"
print(s16.upper())

## lower semua huruf
s17 = "PYTHON IS FUN"
print(s17.lower())

## ganti huruf tertentu
s18 = "banana"
print(s18.replace("a","o"))

## cek apakah string diawali huruf
s19 = "python"
print(s19.startswith("p"))

## cek apakah string diakhiri huruf
s20 = "python"
print(s20.endswith("n"))
```

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
    ‚ÄúString comprehension‚Äù sebenarnya bukan fitur resmi Python seperti list comprehension, tetapi merupakan pola penggunaan list comprehension (atau comprehension lain) untuk mengolah string lalu menghasilkan string baru menggunakan ''.join(...). Karena string itu immutable (nggak bisa diubah langsung), maka Python memberi cara efisien untuk membuat string baru dari proses iterasi dan filter. Dengan teknik ini, kamu bisa memfilter huruf tertentu, memodifikasi huruf, menghapus karakter, sampai melakukan transformasi kompleks dalam 1‚Äì2 baris kode saja. Teknik ini jauh lebih cepat daripada melakukan loop manual yang menggabungkan string satu per satu.

Dalam praktiknya, string comprehension sering digunakan untuk hal-hal seperti menghapus spasi, mengekstrak huruf tertentu, mengubah format string, memproses input pengguna, atau mengamankan data (misal menghapus simbol berbahaya). Selain itu, cara ini juga membuat kode lebih bersih karena rumus logika berada dalam satu baris yang mudah dipahami. Python sendiri sangat mendukung pola ini karena join() sangat efisien untuk membentuk string besar tanpa overhead berlebihan.

FYI: Semua comprehension yang menghasilkan string itu bekerja dengan cara list comprehension ‚Üí lalu digabung dengan "".join(...).

Dengan memahami teknik ini, kamu bisa membuat fungsi pemrosesan teks yang sangat cepat dan ringkas. Bahkan, jika kamu bekerja dengan algoritma text mining, tokenizing, data cleaning, dan NLP dasar, string comprehension adalah teknik wajib yang sering dipakai dalam data preparation.
</div>

```python
text = "a p a   k a b a r "
hasil = "".join([c for c in text if c != " "])
print(hasil)

text = "HeLLo WoRLD"
hasil = "".join([c for c in text if c.isupper()])
print(hasil)

text = "Nomor: 0812-345-678"
hasil = "".join([c for c in text if c.isdigit()])
print(hasil)
```

## ‚öúÔ∏è Function

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
    Function adalah blok kode terorganisir yang dibuat untuk menjalankan tugas tertentu secara berulang tanpa harus menulis ulang kodenya. Python menggunakan konsep modular programming, yaitu memecah program besar menjadi bagian-bagian kecil yang mudah dikelola. Fungsi membantu meningkatkan keterbacaan, mengurangi duplikasi kode, dan mempermudah perbaikan karena kita cukup mengubah logika di satu tempat saja. Function juga dapat menerima parameter sebagai input dan mengembalikan nilai menggunakan return, sehingga bisa digunakan sebagai komponen dalam perhitungan yang lebih besar. Dengan adanya function, kita bisa membuat kode yang lebih ‚Äúreusable‚Äù dan scalable.
</div>

```python
def sapa(nama):
    return f"Halo, {nama}!"
print(sapa("Budi"))

def salam():
    return "Selamat pagi semuanya!"
print(salam())

def tambah(a, b):
    return a + b
print(tambah(5, 7))

def greet(nama="Tamu"):
    return f"Welcome, {nama}!"
print(greet())
print(greet("Rani"))

def luas_persegi(sisi):
    return sisi * sisi
print(luas_persegi(6))

def cetak_data(data):
    print("Isi data:", data)
cetak_data("mantap")

def cek_umur(umur):
    if umur >= 18:
        return "Dewasa"
    return "Anak-anak"
print(cek_umur(20))

def ulang_kata(kata, n):
    for i in range(n):
        print(kata)
ulang_kata("Hallo", 3)

def buat_list(n):
    hasil = []
    for i in range(1, n+1):
        hasil.append(i)
    return hasil
print(buat_list(5))

def jumlah_semua(*angka):
    return sum(angka)
print(jumlah_semua(1, 2, 3, 4, 5))

def info(**data):
    return data
print(info(nama="Eka", umur=20))

def operasi(a, b):
    return a + b, a - b, a * b, a / b
print(operasi(10, 2))

def is_genap(n):
    return n % 2 == 0
print(is_genap(6))

def faktorial(n):
    hasil = 1
    for i in range(1, n+1):
        hasil *= i
    return hasil
print(faktorial(5))

def balik_kata(kata):
    return kata[::-1]
print(balik_kata("python"))

def hitung_vokal(teks):
    vokal = "aiueoAIUEO"
    total = 0
    for huruf in teks:
        if huruf in vokal:
            total += 1
    return total
print(hitung_vokal("Indonesia"))

def max_dari_tiga(a, b, c):
    return max(a, b, c)
print(max_dari_tiga(10, 3, 7))

def tampilkan_range(a, b):
    for i in range(a, b+1):
        print(i)
tampilkan_range(3, 7)

def pangkat(angka, pangkat):
    return angka ** pangkat
print(pangkat(2, 3))

def konversi_celcius_ke_f(c):
    return (c * 9/5) + 32
print(konversi_celcius_ke_f(30))

def is_palindrome(teks):
    return teks == teks[::-1]
print(is_palindrome("level"))
```

## ‚öúÔ∏è Recursive Function

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
    Recursive function adalah fungsi yang memanggil dirinya sendiri sampai mencapai kondisi penghentian (base case). Pendekatan rekursif digunakan untuk menyelesaikan masalah yang dapat dipecah menjadi sub-masalah yang struktur penyelesaiannya mirip, seperti faktorial, fibonacci, tree traversal, dan divide-and-conquer. Rekursi sangat elegan tetapi harus ditulis dengan hati-hati agar tidak memicu infinite recursion atau stack overflow. Python memiliki batas rekursi tertentu, sehingga rekursi dalam Python lebih cocok untuk masalah yang tidak terlalu dalam kedalamannya.
</div>

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
    FYI: Python punya batas kedalaman rekursi bawaan (biasanya 1000 pemanggilan), supaya program tidak crash akibat pemanggilan fungsi yang tak berujung.
</div>

```python
def faktorial(n):
    if n == 1:
        return 1
    return n * faktorial(n-1)
faktorial(5)
```

<div style="background-color:#f5f5f5; border-left:5px solid #555; padding:12px;">
    Meski rekursi terlihat elegan dan bersih, penggunaannya harus hati-hati. Tanpa base case yang jelas, fungsi dapat terjebak dalam lingkaran pemanggilan tak terbatas atau menghabiskan banyak memori stack. Untuk beberapa kasus, pendekatan iteratif (loop) lebih efisien dan cepat. Namun, rekursi tetap menjadi teknik penting dalam pemrograman karena cocok untuk masalah yang memiliki pola pecahan (divide-and-conquer), seperti quicksort, mergesort, dan pencarian pada struktur data yang bercabang. Rekursi bukan hanya teknik, tapi juga cara berpikir dalam memecah masalah.
</div>

```python
def hitung_mundur(n):
    if n == 0:
        return
    print(n)
    hitung_mundur(n-1)

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

def cari(data, target, index=0):
    if index == len(data):
        return False
    if data[index] == target:
        return True
    return cari(data, target, index + 1)

def balik(kata):
    if len(kata) <= 1:
        return kata
    return balik(kata[1:]) + kata[0]

hitung_mundur(5)
print(fibonacci(6))
print(cari([3, 6, 9, 12], 9))
print(balik("kucing"))
```

```python
def jumlah_digit(n):
    if n == 0:
        return 0
    return (n % 10) + jumlah_digit(n // 10)
print(jumlah_digit(12345))

def pangkat_rekursif(a, b):
    if b == 0:
        return 1
    return a * pangkat_rekursif(a, b - 1)
print(pangkat_rekursif(3, 4))

def hitung_huruf(teks):
    if teks == "":
        return 0
    return 1 + hitung_huruf(teks[1:])
print(hitung_huruf("python"))

def sum_list(data, i=0):
    if i == len(data):
        return 0
    return data[i] + sum_list(data, i + 1)
print(sum_list([3, 6, 9, 12]))

def minimum_list(data, i=0):
    if i == len(data) - 1:
        return data[i]
    nilai_min = minimum_list(data, i + 1)
    return data[i] if data[i] < nilai_min else nilai_min
print(minimum_list([7, 3, 8, 2, 9]))

def kelipatan_rekursif(n, k=1, hasil=None):
    if hasil is None:
        hasil = []
    if k > 5:
        return hasil
    hasil.append(n * k)
    return kelipatan_rekursif(n, k + 1, hasil)
print(kelipatan_rekursif(4))

def delete_char(teks, char):
    if teks == "":
        return ""
    sisa = delete_char(teks[1:], char)
    if teks[0] == char:
        return sisa
    return teks[0] + sisa
print(delete_char("banana", "a"))

def faktorial_genap(n):
    if n <= 0:
        return 1
    return n * faktorial_genap(n - 2)
print(faktorial_genap(8))
```

```python
import math

print("Pilih bentuk yang ingin dihitung luasnya:")
print("1. Bujur Sangkar")
print("2. Segitiga")
print("3. Lingkaran")

pilihan = input("Masukkan pilihan Anda (1/2/3): ")

if pilihan == '1':
  try:
    sisi = float(input("Masukkan panjang sisi bujur sangkar: "))
    if sisi < 0:
        print("Error: Sisi tidak boleh bernilai negatif.")
    else:
        luas = sisi * sisi
        print("Luas bujur sangkar adalah:", luas)
  except ValueError:
    print("Input tidak valid, harap masukkan sebuah angka.")

elif pilihan == '2':
  try:
    alas = float(input("Masukkan panjang alas segitiga: "))
    tinggi = float(input("Masukkan tinggi segitiga: "))
    if alas < 0 or tinggi < 0:
        print("Error: Alas dan tinggi tidak boleh bernilai negatif.")
    else:
        luas = 0.5 * alas * tinggi
        print("Luas segitiga adalah:", luas)
  except ValueError:
    print("Input tidak valid, harap masukkan sebuah angka.")

elif pilihan == '3':
  try:
    jari_jari = float(input("Masukkan panjang jari-jari lingkaran: "))
    if jari_jari < 0:
        print("Error: Jari-jari tidak boleh bernilai negatif.")
    else:
        luas = math.pi * (jari_jari ** 2)
        print("Luas lingkaran adalah:", luas)
  except ValueError:
    print("Input tidak valid, harap masukkan sebuah angka.")

else:
  print("Pilihan tidak valid. Anda harus memasukkan angka 1, 2, atau 3.")
```

#### Nomer 2 (Bilangan Ganjil sebanyak n)

```python
try:
    n = int(input("Masukkan jumlah bilangan ganjil yang ingin ditampilkan: "))

    if n <= 0:
        print("Harap masukkan bilangan bulat positif.")
    else:
        print(f"\n{n} bilangan ganjil pertama adalah:\n")
        for i in range(n):
            bilangan_ganjil = 2 * i + 1
            print(bilangan_ganjil)

except ValueError:
    print("Input tidak valid. Harap masukkan sebuah angka.")
```

#### Nomer 3 (Menghitung Deret yang sudah di tentukan)

```python
try:
    total_target = float(input("Masukkan nilai total yang diinginkan: "))

    jumlah_sekarang = 0.0
    n = 1

    print("-" * 30)
    print("Suku-suku deret yang ditampilkan:")
    print("-" * 30)

    while jumlah_sekarang < total_target:
        pembilang = 3 ** n
        penyebut = (4 * n) + 10

        suku_sekarang = pembilang / penyebut
        jumlah_sekarang += suku_sekarang        
        print(f"Suku ke-{n}: {pembilang}/{penyebut} (Nilai: {suku_sekarang:.2f})")
        n += 1
    
    print("-" * 30)

except ValueError:
    print("Input tidak valid. Harap masukkan sebuah angka.")
```

#### Nomer 4 (Menghitung Deret yang sudah di tentukan #2)

```python
try:
    n_suku = int(input("Masukkan jumlah suku yang ingin ditampilkan: "))

    if n_suku <= 0:
        print("Harap masukkan bilangan bulat positif.")
    else:
        print(f"Menampilkan {n_suku} suku deret:")
        penyebut = 14
        for i in range(n_suku):
            # i dimulai dari 0
            pembilang = i + 3
            
            if i % 2 == 0:
                print(f"Suku ke-{i+1}: {pembilang}/{penyebut}")
            else:
                print(f"Suku ke-{i+1}: -{pembilang}/{penyebut}")
            penyebut *= 2

except ValueError:
    print("Input tidak valid, harap masukkan angka.")
```

#### Nomer 5 (Deret matematika pangkat 3)

```python
n = int(input("Masukkan batas maksimal suku (n): "))
jumlah = int(input("Masukkan batas maksimal total penjumlahan (jumlah): "))

suku_sekarang = 3
total_penjumlahan = 0
suku_ke = 1

print("\nDeret:")

while suku_ke <= n and total_penjumlahan + suku_sekarang <= jumlah:
    print(suku_sekarang, end=" ")
    
    total_penjumlahan = total_penjumlahan + suku_sekarang
    
    suku_sekarang = suku_sekarang * 3
    
    suku_ke = suku_ke + 1

print(f"Total penjumlahan akhir: {total_penjumlahan}")
print(f"Jumlah suku yang ditampilkan: {suku_ke - 1}")
```

### Nomer 1 (Tebak Angka)

```python
komp = 6
nyawa = 3
cek_benar = False

while (nyawa > 0):
    user = int(input(f"Masukkan Tebakan ({nyawa} Kesempatan): "))

    if(user < komp):
        print("Terlalu Kecil, naikan angka mu!")
        nyawa -= 1
    elif(user > komp):
        print("Terlalu Besar, turunkan angka mu!")
        nyawa -= 1
    elif(user == komp):
        cek_benar = True
        print("\nSelamat, Jawaban Benar-!")
        break

if(cek_benar == False):
    print(f"\nGame Over, jawabannya adalah: {komp}")
```

### Nomer 2 (Batu, Gunting, Kertas)

```python
import random

list_pilihan = ["gunting", "batu", "kertas"]
komp = random.choice(list_pilihan)
batas_kesempatan = 5
user = str(input("Pilih batu/gunting/kertas:"))
skor_komp = 0
skor_user = 0

while (batas_kesempatan > 1):
    if(user == komp):
        user = str(input("Masukkan Pilihan lagi (SERI):"))
        komp = random.choice(list_pilihan)
        batas_kesempatan -= 1
    elif((user == "batu") & (komp == "kertas")):
        print("Komputer Menang = Anda Batu, Komputer Kertas")
        break
    elif((user == "gunting") & (komp == "batu")):
        print("Komputer Menang = Anda Gunting, Komputer Batu")
        break
    elif((user == "kertas") & (komp == "gunting")):
        print("Komputer Menang = Anda Kertas, Komputer Gunting")
        break
    else:
        print("Selamat Anda Menang-!", "Anda =", user, ", Komputer = ", komp)
        break
```
